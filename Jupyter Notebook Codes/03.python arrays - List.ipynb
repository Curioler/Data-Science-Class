```python
#Python Collections (Arrays): List, Tuple, Set, Dictionary

"""
List is a collection which is ordered and changeable. Allows duplicate members.
Tuple is a collection which is ordered and unchangeable. Allows duplicate members.
Set is a collection which is unordered and unindexed. No duplicate members.
Dictionary is a collection which is unordered, changeable and indexed. No duplicate members.
"""

#List

mylist = ["apple", "banana", "cherry"]
print(mylist)
```

```python
#Access Items (Understanding the index)

print(mylist[1])
```

```python
# Using Negative Index

print(mylist[-1])
```

```python
#Accessing the range (Inclusion/Exclusion)

yourlist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]
print(yourlist[2:5])
```

```python
#Accessing negative range
print(yourlist[-4:-1])
```

```python
#Updating the value
yourlist[1] = "blackcurrant"
print(yourlist)
```

```python
#Looping through a list

newlist = ["apple", "banana", "cherry"]
for x in newlist:
  print(x)
```

```python
#Conditional execution
thislist = ["apple", "banana", "cherry"]
if "apple" in thislist:
  print("Yes, 'apple' is in the fruits list")
```

```python
#Length of a list     

print(len(thislist))
```

```python
#Adding items
thatlist = ["apple", "banana", "cherry"]
thatlist.append("orange")
print(thatlist)
```

```python
#Add item at a certain position (use of method)
somelist = ["apple", "banana", "cherry"]
somelist.insert(1, "orange")
print(somelist)
```

```python
#removing item

oldlist = ["apple", "banana", "cherry"]
oldlist.remove("banana")
print(oldlist)
```

```python
#empty a list

list3 = ["apple", "banana", "cherry"]
list3.clear()
print(list3)
```

```python
#copy a list
list4 = ["apple", "banana", "cherry"]
list5 = list4.copy()
print(list5)
```

```python
#copy using list() method

some_list = ["apple", "banana", "cherry"]
my_list = list(some_list)
print(my_list)
```

```python
#Joining lists

list_1 = ["a", "b" , "c"]
list_2 = [1, 2, 3]

list_3 = list_1 + list_2
print(list_3)
```

```python
#Append list 2 into list 1

list_1 = ["a", "b" , "c"]
list_2 = [1, 2, 3]

for x in list_2:
  list_1.append(x)

print(list_1)
```

```python
#method - extend()

list_1 = ["a", "b" , "c"]
list_2 = [1, 2, 3]

list_1.extend(list_2)
print(list_1)
```

```python
#Constructing a list
this_list = list(("apple", "banana", "cherry")) # note the double round-brackets
print(this_list)
```
